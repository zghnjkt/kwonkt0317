# 데이터 병합하기

```python
adf = pd.DataFrame({'xx' : ['aa', 'bb', 'cc'],
                    'yy' : [1, 2, 3]})
adf
```

```python
bdf = pd.DataFrame({'xx' : ['aa', 'bb', 'dd'],
                    'zz' : ['F', 'T', 'F']})
bdf
```

```python
cdf = pd.DataFrame({'xx' : ['bb', 'cc', 'dd'],
                    'yy' : [2, 3, 5]})
cdf
```

```python
#70. 완쪽 열을 축으로 병합하기
pd.merge(adf, bdf, how = 'left', on='xx')
```

```python
#71. 오른쪽 열을 축으로 병합하기
pd.merge(adf, bdf, how='right', on='xx')
```

```python
#72. 공통 값만 병합하기
pd.merge(adf, bdf, how='inner', on='xx')
```

```python
#73. 모든 값을 병합하기
pd.merge(adf, bdf, how='outer', on='xx')
```

```python
#74. 특정한 열을 비교해서 공통 값이 존재하는 경우만 가져오기
adf[adf.xx.isin(bdf.xx)]
```

```python
#75. 공통 값이 존재하는 경우 해당 값을 제외하고 병합하기
adf[~adf.xx.isin(bdf.xx)]
```

```python
#76. 공통 값이 있는 것만 병합하기
pd.merge(adf, cdf)
```

```python
#77. 모두 병합하기
pd.merge(adf, cdf, how='outer')
```

```python
#78. 어디서 병합되었는지 표시하기
pd.merge(adf, cdf, how='outer', indicator=True)
```

```python
#79. 원하는 병합만 넣기
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge=="left_only"')
```

```python
#80. merge 칼럼 없애기
pd.merge(adf, cdf, how='outer', indicator=True).query('_merge=="left_only"').drop(columns=['_merge'])
```
